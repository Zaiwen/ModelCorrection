```python
terms = list(term_df["term"])
```

```python
def search_ULAN(col):
    result_df = pd.DataFrame(columns=["value", "subject_id", "rel_subject_id",  "rel_term"])
    for person in col:
#         print(person)
        if person.strip() in terms:
            subjectB_id = term_df[term_df["term"] == person.strip()]["subject_id"].values[0]
            subjectA_id = subject_rels_df[subject_rels_df["subjectB_id"] == subjectB_id]["subjectA_id"].values[0]
            rel_term = ";".join((term_df[term_df["subject_id"] == subjectA_id]["term"].values))
            result_df = result_df.append({"value":person.strip(), "subject_id":subjectB_id, "rel_subject_id":subjectA_id, "rel_term":rel_term}, ignore_index=True)
            
        else:
            result_df = result_df.append({"value":person.strip(), "subject_id":0, "rel_subject_id":0, "rel_term":0}, ignore_index=True)
#     result_df.drop_duplicates(inplace=True)
    return result_df
```
