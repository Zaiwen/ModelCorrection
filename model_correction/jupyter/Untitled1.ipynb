```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_excel(r"C:\D_Drive\ASM\experiment\exp_20220419\precisionResult(1 6 12).xlsx")
```

```python
plt.figure(dpi=150)

plt.plot(data["data_source"], data["precision(k>5)"], label="precision(ϕ>5)", marker=">", color="red")
plt.plot(data["recall(k>5)"], label="recall(ϕ>5)", marker=">", ls="--", color="red")
plt.plot(data["precision"], label="precision", marker="*", color="blue")
plt.plot(data["recall"], label="recall", marker="*", ls="--", color="blue")
plt.xticks(fontsize=7)
plt.legend()
```
