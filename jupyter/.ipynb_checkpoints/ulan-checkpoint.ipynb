```python
terms = list(term_df["term"])
```

```python
def search_ULAN(col):
    subject_ids = []
    for person in col:
#         print(person)
        if person.strip() in terms:
            subject_id = term_df[term_df["term"] == person.strip()]["subject_id"].values[0]
            subject_ids.append(subject_id)
        else:
            subject_ids.append(None)
    
#     print(subject_ids)
    return subject_ids
```
